## Visão Geral

Pegue um texto, coloque uma grade ao redor das palavras e então deixe de fora as palavras. O que resta são as posições textuais ou *compartimentos* (*slots* em inglês).

![alt text](https://annotation.github.io/text-fabric/tf/images/DataModel/DataModel.001.png)

Pedaços de um texto correspondem à frases, orações, sentenças, versos, capítulos, livros etc. Desenhe círculos ao redor desses pedaços e então deixe de fora seus conteúdos. O que sobra são objetos textuais, ou seja, *nós* (*nodes* em inglês) ou [*vértices*](https://pt.wikipedia.org/wiki/V%C3%A9rtice_(teoria_dos_grafos)) (para usar a linguagem típica da [teoria dos grafos](https://pt.wikipedia.org/wiki/Teoria_dos_grafos)).

![alt text](https://annotation.github.io/text-fabric/tf/images/DataModel/DataModel.002.png)

Os nós podem ser conectados a outros nós por *arestas* (*edges* em inglês). Uma função básica das arestas é indicar *conteúdo*: *este* vértice (*node*) corresponte a um grupo de *compartimentos* (*slots*) que está contido nos compartimentos *daquele* vértice. Mas as arestas podem também denotar relações mais abstratas e linguísticas entre os vértices.

Vértices, ou nós, têm tipos. Tipos são apenas um rótulo que usamos para fazer distinção entre vértices de palavra, vértices de frase, vértices de livro etc. O tipo designado é um exemplo de um *recurso* (*feature* em inglês) dos vértices: um mapeamento que designa um pedaço de informação para cada vértice. Esse tipo de designação tem um nome: `otype` e todo conjunto de dados do Text-Fabric tem tal recurso.

Vértices podem ser ligados a posições textuais ou compartimentos (*slots*). Alguns vértices são (1) ligados a um único compartimento, outros (2) a um conjunto de compartimentos, e ainda outros (3) a nenhum compartimento.

Vértices do tipo (1) são identificados com seus compartimentos, têm o mesmo número de compartimentos que os seus vértices.

Vértices do tipo (2) têm uma aresta para cada compartimento (que também é um vértice) com o qual eles estão ligados. A coleção destas arestas dos vértices do segundo tipo para vértices do primeiro tipo é um exemplo de um *recurso* (*feature*) das arestas: um mapeamento que designa para cada par de vértices um valor booleano: este par está ligado ou não? Este recurso em particular do vértice é chamado de `oslots` e todo conjunto de dados do Text-Fabric tem tal recurso.

Vértices do tipo (3) representam informação que não é parte do corpo principal do texto. Poderíamos representar o léxico dessa maneira. Contudo, também é possível considera o *lexema* como um tipo de vértice, onde cada vértice lexema está ligado a um conjunto de compartimentos que tem uma ocorrência daquele lexema.

## Metáfora do Tecido

![alt text](https://annotation.github.io/text-fabric/tf/images/loom.png)

[Hausbücher der Nürnberger Zwölfbrüderstiftungen AD 1425](http://www.nuernberger-hausbuecher.de/75-Amb-2-317-4-v/data)

Antes de continuarmos, convidamos você para prestar atenção em alguns termos básicos da arte da [tecelagem](https://pt.wikipedia.org/wiki/Tecelagem).

Um tecelão arruma um conjunto fixo de fios paralelos, o *urdume*. Então ele pega um outro fio, normalmente um colorido, e o passa de maneira perpendicular através do urdume. Este fio é chamado de *trama*.

![alt text](https://upload.wikimedia.org/wikipedia/commons/f/f4/Kette_und_Schu%C3%9F.pt.jpg)

O instrumento que faz atravessar a trama através do urdume é chamado de *tear*. O tecelão continua operando o tear, para frente e para trás, ocasionalmente selecionando novos urdumes, até completar uma peça retangular de tecido, a *malha*.

![alt text](https://annotation.github.io/text-fabric/tf/images/weave.png)
[Fonte](https://sew4home.com/tips-resources/buying-guide/all-about-fabric-weaves-tutorial)

Text-Fabric, a ferramenta, pode ser vista como um terar que envia recursos (urdumes) através de uma trama (o sistema de vértices e arestas).

Os recursos `otype` e `oslots` são aqueles que configuram o sistema de vértices e arestas. É por isso que os chamamos de recursos de *trama*. Cada conjunto de dados do Text-Fabric contém esses dois recursos de trama (mais a frente veremos um terceiro membro da trama, o `otext`). Eles proveem a estrutur de um texto e suas anotações, sem quaisquer conteúdos. O próprio texto é deixado de fora.

![alt text](https://annotation.github.io/text-fabric/tf/images/DataModel/DataModel.003.png)

Todas as outras informações são adicionadas à trama como *recursos* (*features*), as urdiduras: recrusos de vértice e recursos de aresta. Um recurso é um aspecto especial de uma informação textual, isolada como um tipo de módulo. É uma coleção de valores que pode ser tecido como uma urdidora numa trama.

Uma das coisas mais básicas para adicionar à trama é o próprio texto. Textos antigos frequentemente têm várias representações textuais, como os caracteres originais (Unicode) ou os caracteres transliterados, com ou sem o conjunto completo de sinais diacríticos. No Text-Fabric não precisamos escolher entre eles: podemos empacotar cada representação em um recurso e adicioná-lo ao tecido.

Um conjunto de dados do Text-Fabric é uma trama ( `otype`,`oslots`) mais uma coleção de urdumes (todos os outros recursos \[*features*\]). Podemos adicionar outros recursos à mesma trama. Coleções de dados com apenas urdumes mas nenhuma trama são chamados módulos. quando você usa módulos com uma coleção de dados, os módulos devem ser construídos ao redor da trama da coleção de dados.

Quando quer que você use o Text-Fabric para gerar novos dados, você está tecendo a malha. A coleção de dados resultante é um tecido coeso de recursos individuais (urdumes), cujos valores são tomados de uma coleção de nós (trama).

![alt text](https://annotation.github.io/text-fabric/tf/images/DataModel/DataModel.004.png)

Alguns recursos merecem um lugar privilegiado. Afinal, estamos lidando com o *texto*, então precisamos de alguma informação sobre quais recursos carregam representações textuais e informação seccionada (i.e., livros, capítulos, versos).

Esta informação não está imbutida no Text-Fabric, mas é dada na forma de um recurso de configuração (*config*). Um recurso de configuração não tem dados, apenas metadados. Cada conjunto de dados do Text-Fabric pode conter um recurso de configuração chamado `otext` que especifica que tipos de vértices (ou nós) e recursos correspondem a unidades seccionais como livros, capítulos e versos. Também contém modelos para gerar representações de texto para os compartimentos.

O recurso `otext` é opcional, porque não se espera que todas as coleções de dados do Text-Fabric tenham seccionamento extensinvo e definições de representação textual. Especialmente quando você está no processo de converter uma fonte de dados (como um conjunto de dados em formato de árvore \[[*treebank*](https://en.wikipedia.org/wiki/Treebank)\]) em uma coleção de dados do Text-Fabric, é útil que o Text-Fabric possa carregar os dados sem se importar com estas questões.

![alt text](https://annotation.github.io/text-fabric/tf/images/DataModel/DataModel.005.png)

## Modelo

Sumarizamos em breves declarações nosso modelo de dados, incluindo maneira de representar os dados, serializá-los, e computados com ele.

Objetos textuais:
* ocupam composições arbitrárias de compartimento;
* carregam um *tipo* (apenas um rótulo) - todos os *compartimentos* (*slots*) carregam o mesmo tipo, o *tipo slot*, i.e *palavra* ou *caractere*;
* podem ser anotados com *recursos* (*features*), pares de valores-chave;
* podem ser conectos por links diretos, rotulados, com outros objetos textuais.

O modelo sabe que recurso designiu valores a vértices e arestas. Se dois diferentes recursos designam um valor para uma aresta ou vértice, ambos os valores podem ser lidos mais tarde; um através de um recurso e o outro através de outro.

Os dados no Text-Fabric são organizados como um grafo diretamente anotado com um pouco de estrutura adicional. A correspondência é:
* posições textuais => os primeiros tantos números de compartimentos
* objetos textuais => vértices / nós
* ligações entre os objetos textuais => arestas
* informação associada com os objetos textuais => recursos de vértice
* rótulos sobre as ligações entre objetos textuais => recursos de aresta
 * **NB**: uma vez que cada ligação é especificada por um recurso de aresta, cada ligação é implicitamente rotulado pelo nome daquele recurso de aresta. Se um recurso de aresta designa valores para arestas, estes valores vem acima do rótulo implícito.

* tipos de objetos textuais => um recurso de vértice especial chamado `otype` (leia: objeto de tipo)
* extenção dos objetos textuais em termos de compartimentos => um recurso de aresta especial chamado `oslots` (leia: objeto de compartimento)
* especificações opcionais para o seccionamento e representação do texto => um recurso de configuração especial chamado `otext` (leia: objeto de texto)

Juntos, o `otype`, `oslots`, e os recursos opcionais `otext` são chamados de a **trama** de um conjunto de dados do Text-Fabric.

## Representação

Representamos os elementos que criam tal grafo como segue:
* vértices são números inteiros começando com 1, sem saltos;
* os primeiros vértices `maxSlot` correspondem exatamente aos compartimentos, na mesma ordem, onde `maxSlot` é um número de compartimentos;
* vértices maiores que o `maxSlot` correspondem a objetos textuais gerais;
* recursos de vértice são mapeamentos de números integrais para valores;
* recursos de aresta são mapeamentos de pares de números integrais para valores, i.e., arestas são ordenadas em pares de números integrais e arestas rotuladas são [tuplas](https://pt.wikipedia.org/wiki/%C3%89nuplo) ordenadas de dois vértices e um valor;
* valores (para vértices e para arestas) são conjuntos de caracteres (*strings*) em Unicode (utf8) ou números;
* o recurso `otype` mapeia
 * os números inteiros `1..maxSlot` (incluso) para o tipo compartimento, onde `maxSlot` é o último compartimento (*slot*);
 * os números inteiros `maxSlot+1..maxNode` (incluso) para os tipos de objetos textuais relevantes.
* o recurso `oslots` é um recurso de vértice sem valor, mapeando todos os outros vértices que não são compartimentos para coleções de compartimentos - assim há uma aresta `oslots` entre cada vértice que não é compartimento e cada compartimento *contido* por aquele vértice;
* a coleção de dados Text-Fabric é uma coleção de recursos de vértice e recursos de aresta contendo pelo menos os recursos de *trama* `otype`, `oslots` e, opcionalmente, `otext`.

## Mais sobre a Trama

A distinção trama/urdume é uma maneira útil de separar a organização textual do conteúdo textual. Vamos discutir os recursos de trama um pouco mais.

### `otype`: recurso de vértice

Mapeia cada vértice para um rótulo. O rótulo tipicamente é um tipo de objeto que representa um vértice, com valores como
* livro
* capítulo
* verso
* sentença
* oração
* frase
* palavra

Há um tipo especial de tipo de objeto, o tipo compartimento (*slot type*), que é o bloco de construção atômico dos objetos textuais. Se assume que um texto completo é construído a partir de uma sequência de compartimentos (*slots*), do compartimento `1` até o compartimento `maxSlot` (incluso), onde os compartimentos são numerados consecutivamente. Deve haver pelo menos um compartimento.

Todos os outros objetos são definidos com respeito aos compartimentos `slots` que eles contêm.

O tipo compartimento (*slot type*) não precisa ser chamado de *compartimento* (`slot`) literalmente. Se sua entidade básica é `palavra`, você também pode chamá-lo de `palavra`. Compartimentos são então preenchidos com *palavras*. Você pode modelar o texto com base em outra entidade atômica, como um `caractere`. Neste caso, os compartimentos são preenchidos com *caracteres*. Outras escolhas são igualmente viáveis.

O único requerimento é que todos os compartimentos correspondam exatamente ao primeiro de tantos vértices.

O recurso `otype` mapeará o vértice `1` para um típo vértice e este tipo vértice é o tipo de todos os compartimentos subsequentes e também de todas as coisas que preenchem os compartimentos.

Note também que os recursos seccionais `livro`, `capítulo`, `verso` aqui. Eles terão um papel no tercerio, e opcional, recurso de trama, o `otext`.

### `oslots`: recurso de aresta

Define quais compartimentos são ocupados por quais objetos. Ele o faz especificando as arestas dos vértices para os compartimentos que eles contêm.

A partir da informação em `oslots` podemos computar os relacionamentos incorporados entre todos os vértices.

Ele também se ergue ao ordenamento canônico dos vértices (cf. [`tf.core.nodes`](https://annotation.github.io/text-fabric/tf/core/nodes.html)).

### `otext`: recurso de configuração (opcional)

Declara que tipos aresta correspondem aos primeiros três níveis de seccionamento, normalmente *livro*, *capítulo*, *verso*. Também declara os recursos correspondentes para pegar os nomes ou números das seções nestes níveis. O Text-Fabric usa esta informação para construir o assim chamado Text-[API](https://pt.wikipedia.org/wiki/Interface_de_programa%C3%A7%C3%A3o_de_aplica%C3%A7%C3%B5es), com fuções para
* converter vértices para rótulos de seção e vice-versa,
* representar nomes de seção em várias línguas,
* mostrar texto formatado para conjuntos de vértices.
Se não houver informação sobre as seções ou representações textuais, o Text-Fabric construirá um Text-API reduzido para você, mas ele continuará.

## Serialização e Pré-Computação

Quando o Text-Fabric funciona com um conjunto de dados, ele lê arquivos de dados de recurso (*feature data files*) e oferece um API para processar aqueles dados. A principal tarefa do Text-Fabric é tornar o processamento eficiente para que ele possa ser feito de maneiras interativas, como em um Jupyter Notebook. Para este fim, o Text-Fabric
* otimiza o recurso de dados depois de lê-los pela primeira vez e os guarda na forma binária para carregamento rápido em invocações próximas;
* pré-computa dados adicionais dos recursos de trama para prover funções API convenientes.
No Text-Fabric, temos váriás maneiras de codificar este modelo:
* como texto simples em arquivos de recurso [`tf`](https://annotation.github.io/text-fabric/tf/index.html),
* como estruturas de dados Python na memória,
* como serializações comprimidas das mesmas estrutruas de dados dentro de arquivos `.tfx` em diretórios de cache [`.tf`](https://annotation.github.io/text-fabric/tf/index.html).
